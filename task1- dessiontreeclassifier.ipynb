### Task1 - Decision Trees

Lets first import the required libraries ,
matplotlib - popular data visualion library in Python ,
sklearn - popular machine learniibrary built on top of NumPy, SciPy, and Matplotlib ,
numpy -popular Python library used for scientific computing and data analysis .


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
```

lets load the dataset and fit in the model
Dataset - Iris dataset is a well-known machine learning dataset that contains visual measurements of iris flowers
model- A decision tree classifier is a widely used machine learning algorithm in python with the scikit-learn library. It is a supervised learning model that can be used for both classification and regression tasks.


```python
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
print('Decision Tree Classifier Created')
```

Now let us visualize our decision tree using matplotlib -
Once you have created the decision tree object, you can use the 'plot()' method to visualize it using Matplotlib. 
This method will create a graphical representation of the decision tree, showing the decision nodes, leaf nodes, and branching structure.


```python
plt.figure(figsize=(20,10))
plot_tree(dtree, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, fontsize=12)
plt.show()
```

lets make predictions and find the accuracy of the model and aplly some metrics to check model's performance 
An accuracy and classification report is a document that evaluates the performance of a machine learning model. It provides information on the model's overall accuracy, precision, recall, and F1 score. A confusion matrix is a table that shows the number of true positives, true negatives, false positives, and false negatives produced by the model. 

```python
y_pred = dtree.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

```

lets make single prediction using index

```python
sample_index = 0
sample = X_test[sample_index].reshape(1, -1)  # Reshape the sample for prediction
predicted_class = dtree.predict(sample)
predicted_class_name = iris.target_names[predicted_class[0]]

print(f"Sample: {sample}")
print(f"Predicted class: {predicted_class_name}")
```
